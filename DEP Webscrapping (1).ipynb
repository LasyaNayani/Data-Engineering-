```python
import requests
import pandas as pd

# Defining the URL of the NASA dataset (Near-Earth Comets - Orbital Elements)
url = 'https://data.nasa.gov/resource/b67r-rgxc.json'

# Send an HTTP GET request to fetch the JSON data
response = requests.get(url)

# Checking the success of the request
if response.status_code == 200:
    # JSON data
    data = response.json()
    
    # Create a DataFrame from the JSON data
    df = pd.DataFrame(data)
    
    # Display the first 50 records (Start of the index being 0 and end of it being 50)
    print(df.head(50))
    
    # Save the data to a CSV file
    df.to_csv('nasa_comets.csv', index=False)
else:
    print('Failed to retrieve data. Status code:', response.status_code)
```

```python
import pandas as pd

# Read the data from the CSV file
df = pd.read_csv('nasa_comets.csv')

# 1. Handling Missing Values:
# Check for missing values in each column and fill or drop them as needed.
# For example, let's fill missing values in all columns with zero:
df.fillna(0, inplace=True)
```

```python
# 2. Removing Duplicates:
# Check for and remove duplicate rows based on all columns:
df.drop_duplicates(inplace=True)

```
